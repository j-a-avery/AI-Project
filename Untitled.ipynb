```python
import pandas as pd
from PCPartPicker_API import pcpartpicker as pcpp
```

```python
product_types = [
    'cpu', 'cpu-cooler', 'motherboard', 'memory',
    'internal-hard-drive', 'video-card', 'power-supply',
    'case'
]
```

```python
coolers = pcpp.productLists.getProductList('cpu-cooler')
```

```python
coolers
```

```python
coolers_list = [['name', 'fan-rpm', 'noise-level', 'ratings', 'price', 'id']]

for c in coolers:
    if type(c) is dict:
        name, rpm, noise, ratings, price, id = tuple(c.values())
        coolers_list.append([name, rpm, noise, ratings, price, id])
```

```python
coolers_df = pd.DataFrame(coolers_list)

header = coolers_df.iloc[0]
coolers_df = coolers_df[1:]
coolers_df.columns = header

coolers_df
```
